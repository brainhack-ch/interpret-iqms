## 2. PCA analysis 

**Data standarization**

```python
#Remove non numerical data
datatable=datafile.drop(['size_x', 'size_y','size_z','spacing_x', 'spacing_y','spacing_z'], axis=1)

# Separating out the features
features = datatable.columns[1:]
x = datatable.loc[:, features].values

# Separating out the bids_name
y = datatable.loc[:,['bids_name']].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
```

**PCA 2 dimensions**

```python
#Generate 2 components
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
#Generate table with 2 components
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
#Add bids names to pca results table
finalDf = pd.concat([principalDf, datatable[['bids_name']]], axis = 1)
#Plot the PCA

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 10)
ax.set_ylabel('Principal Component 2', fontsize = 10)
ax.set_title('2 component PCA', fontsize = 20)
for i in range(len(finalDf)):
    ax.scatter(finalDf.loc[i,'principal component 1']
               , finalDf.loc[i,'principal component 2']
)
```

**PCA 2 dimension results**

```python
#Variance explained by every component
variance=pca.explained_variance_ratio_
print (variance)
```

```python
#Weight of each variable in component 1

fig= plt.figure(figsize=(20,12))
plt.scatter (features,pca.components_[0])
plt.xticks(rotation=90)
plt.show()   

#Weight of each variable in component 2
fig= plt.figure(figsize=(20,12))
plt.scatter (features,pca.components_[1])
plt.xticks(rotation=90)
plt.show()    
```

***Analysis of the variance explained by n components of the PCA***

```python
pca = PCA()
principalComponents = pca.fit_transform(x)

variancen=(pca.explained_variance_ratio_)
plt.plot(variancen)
```

**PCA with 3 Components**

```python
data=datatable
# Separating out the features
features = data.columns[1:-2]
x = data.loc[:, features].values
# Separating out the bids_name
y = data.loc[:,['bids_name']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
#PCA 2 dimensions
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2','principal component 3'])

#plot
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax = plt.axes(projection ='3d')
ax.set_xlabel('Principal Component 1', fontsize = 10)
ax.set_ylabel('Principal Component 2', fontsize = 10)
ax.set_zlabel('Princpal Component 3', fontsize = 10)
ax.set_title('3 component PCA', fontsize = 20)
for i in range(len(principalDf)):
    ax.scatter(principalDf.loc[i,'principal component 1']
               , principalDf.loc[i,'principal component 2'],
               principalDf.loc[i,'principal component 3']
)
    
```

### 3. PCA separating groups (Scores & Motion)

```python
#loading scores
scores=pd.read_csv('/Users/Usuario/Desktop/BH/scores.tsv', sep='\t', header=0)
```

```python
#Add scores & motion parameters to data table
import re
ratings = []
motion_condition = []
for _, row in datafile.iterrows():
    sub_name = row['bids_name']
    ratings.append(scores.loc[scores['bids_name'] == sub_name, 'score'].item())
    
    patt = re.compile(r"acq-(.*?)_T1w")
    m = patt.search(sub_name)
    if m:
        motion_condition.append(m.group(1))
        
#add them to dataframe        
dfratings = pd.DataFrame(ratings)
dfmotion_condition = pd.DataFrame(motion_condition)
        
finalDf_mc_s= pd.concat([principalDf, datatable['bids_name'], dfratings[0], dfmotion_condition],axis=1)
finalDf_mc_s.columns = ['principal component 1', 'principal component 2', 'bids_name', 'scores', 'movement']

finalDf_mc_s.head()
```

### 4.PCA in just one condition

***Prepare one condition datasets***

```python
#Adding score and movement columns

scmo=pd.concat([dfratings[0], dfmotion_condition],axis=1)
scmo.columns=('scores','movement')
separateDf= pd.concat([datatable, scmo],axis=1)

#Separate data by movement

standard=separateDf[separateDf['movement'] == 'standard']
headmotion1=separateDf[separateDf['movement']== 'headmotion1']
headmotion2=separateDf[separateDf['movement']== 'headmotion2']

#Separate data by score

score1=separateDf[separateDf['scores'] == 1]
score2=separateDf[separateDf['scores']== 2]
score3=separateDf[separateDf['scores']== 3]

standard.head()
```

**Define functions for 2 and 3 components PCA**

```python
#PCA 2c

def PCAanalysis(data,title):    

    # Separating out the features
    features = data.columns[1:-2]
    x = data.loc[:, features].values
    # Separating out the bids_name
    y = data.loc[:,['bids_name']].values
    # Standardizing the features
    x = StandardScaler().fit_transform(x)
    #PCA 2 dimensions
    from sklearn.decomposition import PCA
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2'])

    #plot
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('Principal Component 1', fontsize = 10)
    ax.set_ylabel('Principal Component 2', fontsize = 10)
    ax.set_title(title, fontsize = 20)
    for i in range(len(principalDf)):
        ax.scatter(principalDf.loc[i,'principal component 1']
                   , principalDf.loc[i,'principal component 2']
    )
    
    #plot variables
    fig = plt.figure(figsize = (10,10))
   
    plt.figure(figsize = (10, 30))
    plt.imshow(np.array([pca.components_[0],pca.components_[1]]).T)
    plt.yticks(np.arange(len(pca.components_[0])), features)
    plt.colorbar()
    plt.show()

    #pca 5 components
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(1,1,1) 
    pca = PCA(n_components=5)
    principalComponents = pca.fit_transform(x)
    variancen=(pca.explained_variance_ratio_)
    plt.plot(variancen)

    #ax.legend(y, loc = (1.1,0))
    ax.grid()
    plt.show()
```

```python
#PCA 3c 

def PCAanalysis3(data,title):

    # Separating out the features
    features = data.columns[1:-2]
    x = data.loc[:, features].values
    # Separating out the bids_name
    y = data.loc[:,['bids_name']].values
    # Standardizing the features
    x = StandardScaler().fit_transform(x)
    #PCA 2 dimensions
    from sklearn.decomposition import PCA
    pca = PCA(n_components=3)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2','principal component 3'])

    #plot
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(1,1,1) 
    ax = plt.axes(projection ='3d')
    ax.set_xlabel('Principal Component 1', fontsize = 10)
    ax.set_ylabel('Principal Component 2', fontsize = 10)
    ax.set_zlabel('Princpal Component 3', fontsize = 10)
    ax.set_title(title, fontsize = 20)
    for i in range(len(principalDf)):
        ax.scatter(principalDf.loc[i,'principal component 1']
                   , principalDf.loc[i,'principal component 2'],
                   principalDf.loc[i,'principal component 3']
    )
        
        #plot variables
    fig = plt.figure(figsize = (10,10))
   
    plt.figure(figsize = (10, 30))
    plt.imshow(np.array([pca.components_[0],pca.components_[1],pca.components_[2]]).T)
    plt.yticks(np.arange(len(pca.components_[0])), features)
    plt.colorbar()
    plt.show()

```
